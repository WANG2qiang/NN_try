# 1-神经网络的数学模型以及相关基本概念介绍

人工神经网络这个概念是在对人类中枢神经系统的观察中受到的启发。

因此在我们开始介绍人工神经网络的特点之前，我们先看看自然界中高等动物的神经系统有哪些（结构以及功能上的）特点：

（1）由大量基本结构和功能单位（神经元）相互连接组成（树突、轴突）；

（2）单个“神经元”具有感受刺激、整合信息和传导冲动的能力（“神经元”感知环境的变化后，再将信息传递给其他的“神经元”）；

（3）单个“神经元”基本都包含接收（receptive zone）、触发（trigger zone）、传导（conducting zone），和输出（output zone）四个功能区

（4）单个“神经元”的功能相对（整个系统而言）较为简单；

（5）单个“神经元”的功能一般很难描述（全有全无律，all-or-none law）

（6）整个“神经系统”能够实现及其复杂的功能（思考、判断、推理）。

以上的特点可能还不够全面，但是已经总结出了很多人工神经网络与中枢神经系统的相似性，我们将根据这些相似性来理解人工神经网络。

中枢神经系统虽然复杂，但是通过研究组成它的基本单元，我们仍然能够构建起神经科学，总结出神经系统的各种规律。

同样的，研究人工神经网络时，我们先从人工神经元（**Artificial Neuron**）开始。

P.S. 关于全由全无率（all-or-none law）是骨骼肌或心肌的肌纤维或是神经反应独立于刺激的强度。如果刺激超过阈电位，神经或肌纤维就会有对应反应；否则，就完全不会有反应。

### 人工神经元（Artificial Neuron）

一个典型的人工神经元示意图如下图所示



人工神经元通过多个输入通道感受刺激，通过求和的方式整合信息，利用激活函数处理和传导信息，其功能和神经元细胞的功能存在一定相似性。

矩阵乘法我们不需要过多讨论，而激活函数（**Activate function**）很明显是人工神经元的核心。

由于神经元的全有全无律（**all-or-none law**）特性，激活函数一般都被假定为非线性函数。

所以，我们在这里先简单介绍一些常用于实现人工“神经元”的非线性函数（以及它们的导数）。

这些函数以及它们的导数在我们后面构建神经元以及神经网路时可能用得上。

### 人工神经元的激活函数（Activate function）

激活函数（**Activate function**）的作用类似生物神经元中的动作电位，

由于神经元功能上的全有全无律，激活函数一般是非线性函数。

常用的激活函数主要有下面几种：

Rectified Linear Unit(ReLU)函数：

$$
h(a) = max\left\{{0,a}\right\}
$$

ReLU函数的导数：

$$    
h'(a) =
    \left\{
        {\begin{array}{cc}
             0, & for\space a<0 \\
             1, & for\space a>0 \\
         \end{array} }
    \right.
$$

Leaky ReLU函数:

$$    
h(a) =
    \left\{
        {\begin{array}{cc}
             \alpha a, & for\space a<=0 \\
                  a, & for\space a>0 \\
         \end{array} }
    \right.
$$

Leaky ReLU函数的导数：

$$    
h'(a) =
    \left\{
        {\begin{array}{cc}
             \alpha, & for\space a<=0 \\
               1  , & for\space a>0 \\
         \end{array} }
    \right.
$$

Sigmoid函数/S形函数/Logistic函数:

$$    
h(a) = \frac{1}{1 + e^{-a}}
$$

Sigmoid函数/S形函数/Logistic函数的导数：

$$    
h'(a) = h(a)(1-h(a)) = \frac{e^{-a}}{(1 + e^{-a})^2}
$$

Softmax（归一化指数）函数：

$$
h(\mathbf{a}) = \frac{e^{a_k}}{\sum_{k=1}^{K}{e^{a_k}}} 
    = 
        \left[
            \frac{e^{a_1}}{\sum_{k=1}^{K}{e^{a_k}}},
            \frac{e^{a_2}}{\sum_{k=1}^{K}{e^{a_k}}},
            \cdots 
            \frac{e^{a_K}}{\sum_{k=1}^{K}{e^{a_k}}}
        \right]
    =
        \left[
            h_1,
            h_2,
            \cdots 
            h_K
        \right]
$$

由于Softmax函数返回的是一个向量，因此求导会得到一个矩阵，还记得我们上一节课讲到的矩阵求导吗？

根据矩阵求导规则，我们可以得出Softmax（归一化指数）函数的导数:

$$
\frac{\partial(h(\mathbf{a}))}{\partial{\mathbf{a}}} =
\begin{array}{c|l}
{↓}&{h_1}&{h_2}&{h_3}&{\cdots}&{h_K}\\
\hline
{a_1}   
&{\frac{\partial{{h_1}}}{\partial{a_1}}}
&{\frac{\partial{{h_2}}}{\partial{a_1}}}
&{\frac{\partial{{h_3}}}{\partial{a_1}}} 
&{\cdots}
&{\frac{\partial{{h_K}}}{\partial{a_1}}} 
\\
{a_2}   
&{\frac{\partial{{h_1}}}{\partial{a_2}}}
&{\frac{\partial{{h_2}}}{\partial{a_2}}}
&{\frac{\partial{{h_3}}}{\partial{a_2}}} 
&{\cdots}
&{\frac{\partial{{h_K}}}{\partial{a_2}}} 
\\
{a_3}   
&{\frac{\partial{{h_1}}}{\partial{a_3}}}
&{\frac{\partial{{h_2}}}{\partial{a_3}}}
&{\frac{\partial{{h_3}}}{\partial{a_3}}} 
&{\cdots}
&{\frac{\partial{{h_K}}}{\partial{a_3}}} 
\\
{\cdots}&{\cdots}&{\cdots}&{\cdots}&{\cdots}&{\cdots}  
\\
{a_K}   
&{\frac{\partial{{h_1}}}{\partial{a_K}}}
&{\frac{\partial{{h_2}}}{\partial{a_K}}}
&{\frac{\partial{{h_3}}}{\partial{a_K}}}
&{\cdots}
&{\frac{\partial{{h_K}}}{\partial{a_K}}} 
\\
\end{array}
$$

所以有：

$$
\frac{\partial(h(\mathbf{a_i}))}{\partial{\mathbf{a_j}}} = (h(\mathbf{a}_i))(\delta_{ij} - h(\mathbf{a}_j))
$$

其中 $ \delta_{ij} $ 是克罗内克 $ \delta $ 函数：

$$    
\delta_{ij} =
    \left\{
        {\begin{array}{cc}
             0, & i \neq j \\
             1, & i = j \\
         \end{array} }
    \right.
$$



Softmax函数是二分类函数sigmoid在多分类上的推广，目的是将多分类的结果以概率的形式展现出来。

大家可以自己举一个简单的例子来测试一下看看。

TanH(双曲正切)函数：

$$
h(a) = \frac{e^{a} - e^{-a}}{e^{a} + e^{-a}}
$$

TanH(双曲正切)函数的导数：

$$
h'(a) = (1 - h^2(a)) = \frac{4}{e^{2a} + e^{-2a} + 2}
$$

把这些函数的图像画出来，我们可以发现这些函数多多少都有一点全有全无律（all-or-none law）的影子。

P.S.虽然不是所有的神经元动作都遵循全有全无律，但是这一特点显然被最为广泛的应用着，所以我们一般也都会采用上面的函数来实现神经网络。

```python
# 画函数图像
# 导入库
import numpy as np
import matplotlib.pyplot as plt

# 创建数组
x = np.linspace(-10,10,100)

# 定义函数
def RelU(input):
    if input <0:
        output = 0
    else:
        output = input
    return output

def Sig(input):
    output = 1/(1 + np.exp(-input))
    return output

def tanh(input):
    output = (np.exp(input) - np.exp(-input))/(np.exp(input) + np.exp(-input))
    return output

# 画图
y1 = np.zeros(x.shape[0])
y2 = np.zeros(x.shape[0])
y3 = np.zeros(x.shape[0])
for i in range(x.shape[0]):
    y1[i] = RelU(x[i])
    y2[i] = Sig(x[i])
    y3[i] = tanh(x[i])
    
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)
plt.show()
```

### 人工神经元的功能描述

有了上面的这些激活函数，再加上矩阵乘法，我们就构建人工神经元的数学模型了。

通过构建人工神经元的数学模型来描述人工神经元如何工作（传导信号）

仍然以图示的神经元为例（虽然在图中用的是 $ f $ 来表示激活函数，而我们在课件中用的是 $ h $ 但是符号不代表本质，大家清楚是激活函数就可以了）：



那么这个人工神经元的输入输出之间就可以通过这样的式子来描述：

$$
t_{output} = f(z) = f(a_1*w_1 + a_2*w_2 + a_3*w_3 + \cdots + b) = f(\mathbf{w^T*a} + b)
$$

其中

$$
z = \mathbf{w^T*a} + b = a_1*w_1 + a_2*w_2 + a_3*w_3 + \cdots + b
$$

虽然我们还没有讲解什么是神经网络的“正向传播”与“反向传播”，但是我们这里还是能看出来，上面的公式给出了神经元怎么把输入信号转换为输出的过程，也就是“正向”的过程。

既然有“正向”那么肯定有“反向”，不过这里的“反向”并不是要给出反函数，整个 $ f(z) $ 相当于一个多元函数，也不太好算出反函数来。

假设我们给定了一组输入 $ \mathbf{a} $ 和权重系数 $ \mathbf{w},b $ 那么我们就有了输出

$$
t_{output} = f(z)
$$

这个输出和我们想要的真实值 $ y_{gt} $ 之间大概率存在一定误差，而我们肯定希望尽可能地减小误差。

我们用 $ L^2 $ 范数来衡量这个误差，那么就有

$$
\varepsilon = \frac{1}{2}[f(z) - y_{gt}]^2
$$

由于输入 $ \mathbf{a} $ 已经给定，我们只能通过调节权重系数 $ \mathbf{w},b $ 来控制误差，所以我们需要求出 $ \varepsilon $ 对每个权重系数的偏导数，来通过梯度下降法来“搜索”最佳的$ \mathbf{w},b $

那么就有
$$
\begin{aligned}
\frac{\partial\varepsilon}{\partial{w_i}} 
& = \frac{\partial\varepsilon}{\partial{z}}\frac{\partial{z}}{\partial{w_i}} \\
& = \frac{\partial\varepsilon}{\partial{z}}*a_i \\
\end{aligned}
$$

当然，这其实是后面我们讲解神经网络时的内容，不过既然神经网络是很多神经元组成的，所以我在这里从就先从神经元开始涉及这部分概念，为后面的讲解做一些准备。

我在这里的理解是，“反向”给出了神经元的不同权重系数对输出的影响情况。

#### 拓展-双曲函数简介以及其他双曲函数

在数学中，双曲函数是一类与常见的三角函数类似的函数，它们包括：

双曲正弦：
$$
sinh(x) = \frac{e^{x} - e^{-x}}{2}
$$

双曲余弦：
$$
cosh(x) = \frac{e^{x} + e^{-x}}{2}
$$

双曲正切：
$$
tanh(x) = \frac{sinh(x)}{cosh(x)} = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}
$$

双曲余切：
$$
coth(x) = \frac{cosh(x)}{sinh(x)} = \frac{e^{x} + e^{-x}}{e^{x} - e^{-x}}
$$

双曲正割：
$$
sech(x) = \frac{1}{cosh(x)} = \frac{2}{e^{x} + e^{-x}}
$$

双曲余割：
$$
csch(x) = \frac{1}{sinh(x)} = \frac{2}{e^{x} - e^{-x}}
$$

如同点 $ (cos(t),sin(t)) $ 定义一个圆，点 $  (cosh(t),sinh(t)) $ 定义了右半直角双曲线 $ x^2- y^2= 1 $。

带入复数 **(complex number)** 可以很容易的发现双曲函数与三角函数之间的关系，这里不罗嗦了。

### 从神经元到神经网络

构造完了神经网络的基本单位“神经元”，下面我们就要构造神经网络了。

神经网络是由神经元互相连接形成的，尽管实际上可能存在的连接方式有无数种，但是我们最常用的，或者说最常讨论到的连接方式

一般都是下图这种：



按照“层”来进行。

这种方式足够清晰明了，而且非常有代表性，其他连接方式的网络通过一定的拓扑变形（就像化简电路图那样）最终都可以变形成这种结构。

下面来对整个网络进行数学上的描述。

单个神经元传输情况的数学表述我们已经知道了，而具有相同激活函数的神经元组成了一个又一个“层”。

我们先给出某一“层”的表述，再通过层之间组成网络。

对于任意一层（比如第 $ l $ 层），我们假设这一层的输入为 $ \mathbf{a_l} $，权重系数为 $ \mathbf{W_l, b_l} $，这一层的激活函数都是$ \mathbf{h_l(z)} $，输出为$ \mathbf{o_l} $

那么就有

$$
\mathbf{o_l} = \mathbf{h_l(z)} = \mathbf{h_l(W_{l}^{T}*a_l + b_l)}
$$

而这一层的输出是下一层的输入，这一层的输入是上一层的输出，所以假设有一个如图所示的$ L $层神经网络



那么我们可以写出整个网络的函数为：

$$
\mathbf{f(x) = h_L(W_L(h_{L-1}(W_{L-1}\cdots h_1(W_1x))))}
$$

这里为了形式简洁，就把偏置系数$ \mathbf{b} $一起并入了权重矩阵 $ \mathbf{W} $，我们在线性回归那节课也用到了类似的操作。

熟悉数学的同学可以看出，这有点类似于数学上的叠函数（迭代函数）问题，这类问题可以很简单，比如：

$$
f(f(x)) = x
$$

也可以很复杂，比如：

$$
f(f(x)) = sin(x)
$$

这类问题通常很难直接求解，当然，我们在这里不需要去求解这一类纯数学问题。

我们需要的，是这个神经网络能否实现我们想要的功能，如果神经网络的输出结果和我们预期的有偏差，那么怎样才能不断地修正这个神经网络让它达到我们想要的结果。

仍然以上面的$ L $层神经网络为例。

这个神经网络具体有多少层，每一层又有多少个节点是可以根据情况自行决定的。

但是不管怎样，我们的最终目标都是一样的，都是希望误差尽可能地小。

在初始化一个网络的时候，这个网络的参数我们是随便乱给的，因为并不知道每个权重参数应该给多少。

所以当训练集给定的时候必然会有误差，而这些误差是由于权重参数引起的。

用误差函数来表现，就是：

$$ 
\begin{aligned}
{\varepsilon} 
&= \mathbf{{\varepsilon}(W_1, W_2, \cdots, W_L)} \\
&= \frac{1}{2}\mathbf{(y - y_{gt})^T * (y - y_{gt})}\\
&= \frac{1}{2}\mathbf{(h_L(a_L) - y_{gt})^T * (h_L(a_L) - y_{gt})}\\
\end{aligned}
$$

显然，让误差函数最小，我们需要误差函数对权重参数的导数为0。

不过直接求解这么多参数的方程组显然太麻烦了，所以一般都用梯度下降法来求解，而要应用梯度下降法，我们就必须得知误差函数针对每个参数的偏导数。

为了追求一般性，我们先考虑神经网络中任意的一层 $ l $ ，其中 $ w_{ij}^l $ 表示第 $ l $ 层中从第 $ j $ 个输入向第 $ i $ 个节点（神经元）传递数据时的权重系数。



我们就来求一下误差函数 $ {\varepsilon} $ 对 $ w_{ij}^l $ 的偏导数，那么有

$$
\frac{\partial{\varepsilon}}{\partial{w_{ij}^l}} 
= \frac{\partial{\varepsilon}}{\partial{a_i^l}} * \frac{\partial{a_i^l}}{\partial{w_{ij}^l}}
= \frac{\partial{\varepsilon}}{\partial{a_i^l}} * \frac{\partial{(w_{ij}^l * o_j)}}{\partial{w_{ij}^l}}
= \frac{\partial{\varepsilon}}{\partial{a_i^l}} * o_j
$$

而 $ a_i^l $ 是这一层第 $ i $ 个节点接收到的总输入，通过节点 $ i $ 的激活函数 $ o_i^l = h_i^l(a_i^l) $ 向下一层传输数据，影响误差。

所以有

$$
\frac{\partial{\varepsilon}}{\partial{a_i^l}}
= \frac{\partial{\varepsilon}}{\partial{o_i^l}}*\frac{\partial{o_i^l}}{\partial{a_i^l}}
= (\sum{\frac{\partial{\varepsilon}}{\partial{a_k^{l+1}}}*\frac{{\partial{a_k^{l+1}}}}{\partial{o_i^{l}}}})*(h_i^{l'}(a_i^l))
$$

这就和上一层发生了联系，而到了最后一层，也就是和输出直接相连的第 $ L $ 层，我们有

$$
\frac{\partial{\varepsilon}}{\partial{a_i^L}} = (h_i^L(a_i^L) - y)*(h_i^{L'}(a_i^L))
$$

所以这种从最后一层的偏导数开始，逐渐向钱求出每一层的偏导数的方式，就是神经网络中经常提到的“反向传输”。

我们的“反向传输”需要根据上一层传过来的偏导数值

$$
\frac{\partial{\varepsilon}}{\partial{a_i^{l+1}}}
$$

以及这一层的导数值

$$
h_i^{l'}(a_i^l)
$$

来计算出这一层的偏导数值

$$
\frac{\partial{\varepsilon}}{\partial{a_i^l}}
= \frac{\partial{\varepsilon}}{\partial{o_i^l}}*\frac{\partial{o_i^l}}{\partial{a_i^l}}
= (\sum{\frac{\partial{\varepsilon}}{\partial{a_k^{l+1}}}*w_k^{l+1}})*(h_i^{l'}(a_i^l))
$$

再根据篇导数值得出针对不同权重系数的偏导数（数值），以便用梯度下降法更新权重

$$
W_l = W_l - lr * \frac{\partial\varepsilon}{\partial{W_l}} = W_l - lr * \frac{\partial\varepsilon}{\partial{a_l}}*o_{l-1}
$$

通过反向传输，我们能够算出误差函数对于每个权重系数的偏导数，有了偏导数，我们就能使用梯度下降的方法不断优化。

我们需要不断地进行正向传输与反向传输来不断地修正权重系数，直到误差足够小。

在这个范例里面我们构造了一个只有1层（只有一个隐藏层）的神经网络，我们试图用这个网络去进行线性拟合。

程序如下：

```python
# 导入库
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt


# 通过class关键字定义一个神经网络的类
# 包含前向和损失函数，由于采用了PyTorch包，比较麻烦计算的后向解就不需要我们手动计算了。
class NeuralNet(nn.Module):
    # 初始化部分，定义了层数，每层多少个结点，以及采用的激活函数是什么类型
    # 如果要增加层数，或者修改激活函数，可以在这里进行
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 4),
            nn.ReLU(),
            nn.Linear(4, 1)
        )
        
    # 前向解（后向解可以通过网络自动计算）
    def forward(self, x):
        return self.net(x)
    
    # 损失函数，指定使用MSE作为损失函数
    def cal_loss(self, pred, target):
        criterion = nn.MSELoss(reduction='mean')
        return criterion(pred, target)

    
if __name__ == '__main__':
    # 创建数据集
    x = np.arange(0, 1, 0.01).reshape(100, 1)
    y = x * 10 + 10
    x, y = torch.tensor(x).float(), torch.tensor(y).float()
    # 根据NeuralNet类创建神经网络
    model = NeuralNet()
    # 对网络参数进行优化，指明学习率
    optim = torch.optim.SGD(model.parameters(), lr=0.001)
    n_epoch = 100
    epoch = 0
    # 开始训练
    while epoch < n_epoch:
        # 将model转换为训练模式
        model.train()
        for i in range(len(x)):
            optim.zero_grad()
            pred = model(x[i])
            loss = model.cal_loss(pred, y[i])
            loss.backward()
            optim.step()
            # print(loss.detach().cpu().item())
        epoch += 1
    # 将model转换为评估模式
    model.eval()
    # 输出结果
    for i in range(len(x)):
        print('x:{} y:{} y^:{}'.format(x[i], y[i], model(x[i]).item()))
```

正式开始之前，我们再梳理一下全连接神经网络的编程流程。

神经网络的编程流程主要有下面几个步骤：

（1）载入训练集

（2）给定一组初始化权重参数$ \mathbf{W} $（一般用随机数来实现）；

（3）随机选取一个或者一组数据点（一般是一个，每次循环都要重新选取）；

（4）根据这个选定的数据点，利用已经定义好的每层的激活函数，以及初始化权重参数$ \mathbf{W} $，计算出每一层的输入与输出（正向传递/正向解）；

（5）计算到最后一层时，再反过来，从最后一层向前计算误差项与每一层输入之间的梯度

（6）利用梯度下降法更新权重参数$ \mathbf{W} $

（7）重复**step（3） - step（6）**的过程，直到误差足够小。

（8）训练完成，载入测试集进行测试。

我们利用 **class** 关键字，一步一步的实现。

我们首先选取一个非常简单的例子。

输入只有$ x $，然后经过线性变换 $ [w,b] $ 最终输出 $ y = w*x + b $

也就是
$$
x \space \rightarrow \space \left[\ [w,b] \right]\ \space \rightarrow \space y = w*x + b
$$

我们不妨设
$$
x = 1.5 \space y_{tr} = 0.8
$$

而参数的初始值我们设为
$$
w = 0.8 \space b = 0.2
$$

首先我们定义激活函数 **（activate function）**

显然，这个过于简单的范例，激活函数是一个线性函数

```python
# 导入库
import numpy as np


# 定义激活函数
class Activator(object):
    # 前向传播（权重后的输入如何转化为输出）
    def forward(self, weighted_input):
        self.weighted_input = weighted_input
        self.output_data = self.weighted_input
        return self.output_data

    # 后向传播（导数，可以根据输入权重后的输入计算，也有根据输出来计算的案例）
    # 我这里为了方便理解，采用了数学上常见的以输入作为自变量的求导形式
    def backward(self, weighted_input):
        self.weighted_input = weighted_input
        return 1.0
```

然后是全连接层，我们这里只有一层

```python
# 定义全连接层
class FullConnectLayer(object):
    def __init__(self, input_size, output_size, activator):
        # self.input_size：输入维度
        # self.output_size：输出维度
        # self.activator：激活函数
        self.input_size = input_size
        self.output_size = output_size
        self.activator = activator
        # 权值矩阵self.w初始化
        # 权值矩阵self.b初始化
        self.w = np.array([0.8])
        self.b = np.array([0.2])
        self.w_grad_total = np.zeros((output_size, input_size))
        self.b_grad_total = np.zeros((output_size, 1))
        # 后续自定义方法（def）中的参数规范化

    # 正向计算过程
    # 根据最开始的原始输入（input）逐渐计算应该向下一层输出（output）的数据。
    def forward(self, input_data):
        self.input_data = input_data
        self.output_data = self.activator.forward(np.dot(self.w, self.input_data) + self.b)
        return self.output_data

    # 反向计算过程
    def backward(self, input_delta):
        self.input_delta = input_delta
        self.sen = input_delta * self.activator.backward(self.input_data)  # 这一层的偏导数（对权重整合后的输入）
        output_delta = np.dot(self.w.T, self.sen)  # 这一层的偏导数和权重相乘（下一层的误差）
        self.w_grad = np.dot(self.sen, self.input_data.T)  # 对这一层的某一个具体的 w 的偏导数
        self.b_grad = self.sen
        self.w_grad_total = self.w_grad_total + self.w_grad
        self.b_grad_total = self.b_grad_total + self.b_grad
        return output_delta

    def update(self, lr):
        self.w = self.w - lr * self.w_grad
        self.b = self.b - lr * self.b_grad
```

之后定义神经网络

```python
# 定义神经网络
class NeuroNetwork(object):
    def __init__(self, params_array, activator):
        # params_array为层维度信息超参数数组
        # layers为网络的层集合
        self.layers = []
        for i in range(len(params_array) - 1):
            self.layers.append(FullConnectLayer(params_array[i], params_array[i + 1], activator))

    # 网络前向迭代
    def predict(self, sample):
        # 下面一行的output可以理解为输入层输出
        output = sample
        for layer in self.layers:
            layer.forward(output)
            output = layer.output_data
        return output

    # 网络反向迭代
    def calc_gradient(self, label):
        delta = (self.layers[-1].output_data - label)
        for layer in self.layers[::-1]:
            delta = layer.backward(delta)
        return delta

    def update(self, lr):
        for layer in self.layers:
            layer.update(lr)

    def loss(self, pred, label):
        return 0.5 * ((pred - label) * (pred - label)).sum()

    # 一次训练一个样本 ，然后更新权值
    def train_one_sample(self, sample, label, lr):
        self.predict(sample)
        Loss = self.loss(self.layers[-1].output_data, label)
        self.calc_gradient(label)
        self.update(lr)
        return Loss
```

然后带入一组数据看看

```python
params = [1, 1]
activator = Activator()
net = NeuroNetwork(params, activator)

data = np.array([[1.5]])
label = np.array([[0.8]])
for i in range(30):
    print(net.predict(data))
    loss = net.train_one_sample(data, label, 0.1)
    print('loss: %f' % loss)
```

# 关于大作业

我们差不多还有两次课就结课了，这里跟大家说一下我们的大作业。

大作业的题目一共有两个课题，大家可以选择其中之一。

题目一：脑部MRI图像的自动配准

问题描述：

医学图像配准是医学图像分析中的一项基础且重要的任务，而脑部 MRI 图像配准是衡量医学图像配准算法性能的一个普遍任务。

由于脑部图像的解剖结构复杂多变，因此配准脑部 MRI 图像是一项具有挑战性的任务。

传统的配准算法由于计算复杂度高，往往需要耗费较多的时间。

基于深度学习的配准方法由于其计算速度快，在近年内得到了广泛的发展。

然而，现有的脑部 MRI 图像配准方法常常忽略配准后解剖性质的保留。

因此，在保证解剖性质不变的前提下，如何实现高精度的配准是该领域亟待解决的难题。

通过所提供的的数据集，要求各位同学实现一个神经网络模型，能够实现两套三维脑部 MRI 图像之间的配准。

题目二： 新冠肺炎肺部 CT 图像分割

问题描述：

医学图像分割是计算机辅助诊断的关键步骤，其目的是使图像中解剖或病理结构的变化更加清晰，以提高诊断的效率和准确性。

目前，新冠肺炎肺部的病灶分割仍然依靠人工在 CT 图像上进行手动勾画，勾画精度取决于医师的经验和熟练程度，存在很大的主观性；此外，手动勾画速度慢，人力成本高，医生工作负担大。

因此，采用计算机实现鼻咽癌关键器官自动勾画已成为必然趋势。

通过所提供的数据集，要求各位同学实现一个神经网络模型，实现病人 CT 图像中肺部病变的分割。

该数据集包含20个标记的新冠肺炎CT数据，包含左肺、右肺和感染区域都经过专业人员标记和验证。

这些题目是从《2021年全国大学生生物医学工程创新设计竞赛》中选取的，可能存在一定的难度，建议各位同学可以2-4人组成一组，以组为单位提交。

提交报告的形式仍然用jupyter notebook，然后说明小组成员，以及每个组员所负责的部分工作。
